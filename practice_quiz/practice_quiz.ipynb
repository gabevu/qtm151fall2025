### c) For-loops + Plotting

   - Create an array of x-values with 100 evenly spaced points between $0$ and $2\pi$ <br>using `np.linspace()`. **Hint**: type ``help(np.linspace)`` in a separate cell to see how it is used

   - Create a vector named ```vec_y1``` where each element is computed from the formula <br> $ y_k = \sin(x_k)$. (Here $x_k$ is the $k$-th element of the array you created above, and $y_k$ is the $k$-th element of vec_y1)

   - Create a vector ```vec_y2``` where each element is computed from the formula <br> $ y_k = \cos(x_k)$.

   - Create a list ```list_vectors``` with elements ```vec_y1``` and ```vec_y2```. (So a list with two components, both being numpy arrays)

   - Create a for-loop that goes through each element of ```list_vectors``` and plots each <br>list of $y$ values on the vertical axis against $x$ on the horizontal axis.

Note: The final output should be a 2-dimensional plot with the values from vec_y1 and vec_y2 on the vertical axis, and the $x$ values on the horizontal axis. **Remember to label the axes and add a title to the plot**.

````python
# Write your answer here:
import numpy as np
import matplotlib.pyplot as plt
x_values = np.linspace(0, 2*np.pi, 100)
vec_y1 = np.sin(x_values)
vec_y2 = np.cos(x_values)
list_vectors = [vec_y1, vec_y2]

for vector in list_vectors:
    plt.plot(x_values, vector)

plt.title("Graph 1: Sine and Cosine Waves between 0 and 2pi")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()


````

Generate random samples from two specified distributions and plot their histograms, side by side in a single figure. Each sub-figure will have its own labeled axes and title.



````python
#answer here:

#import anything you'll need to set up the histograms.
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd

#Make fake samples to use in our histograms 
normal_sample = np.random.normal(loc=0, scale=1, size=1000)
uniform_sample = np.random.uniform(low=-3, high=3, size=1000) #low is min, high is max

#set up the subplots
fig, subfig_array = plt.subplots(nrows = 1, ncols = 2, figsize= (6, 3))

#set up the first histogram
subfig_array[0].hist(normal_sample, color='blue', edgecolor='black')
subfig_array[0].set_title("Graph 4: Normal Distribution")
subfig_array[0].set_xlabel("Value")
subfig_array[0].set_ylabel("Frequency")
#Set up the second histogram
subfig_array[1].hist(uniform_sample, color='orange', edgecolor='black')
subfig_array[1].set_title("Graph 5: Uniform Distribution")
subfig_array[1].set_xlabel("Value")
subfig_array[1].set_ylabel("Frequency")

#adjust x-axis
plt.xlim(-4, 4)

#tight layout
plt.tight_layout()
plt.show()

````

Create a list of .csv file names and an additional empty list. Then in a for loop, read in each file to make a DataFrame, compute the mean of one of the variables/features/columns, and store it as an element of the originally empty list. Then print the list. 



````python
#import
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#create a csv file of the existing data
csv_files = ["practice_quiz_data/korea_rep_co2_emissions.csv", "practice_quiz_data/united_states_co2_emissions.csv"]

#set up an empty list 
dataframes = []

#create the for loop to read each csv file and append it to the empty list 
for file in csv_files: 
    df = pd.read_csv(file)
    mean_value = df["total_emissions"].mean()
    dataframes.append(mean_value)

print(dataframes)
````
